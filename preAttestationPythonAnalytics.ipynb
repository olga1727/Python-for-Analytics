## 1.2.**Изучение стоимости недвижимости**

```python
prices = df['price'].values
n_bins = 10
plt.figure(figsize=(7, 9))
plt.hist(prices, bins=n_bins)
plt.xlabel('Стоимость недвижимости')
plt.ylabel('Количество недвижимости')
plt.title('Распределение цен на недвижимость')
plt.show()
```

Чем выше чек, тем меньше объектов недвижимости представлено на рынке.

1.3. **Изучите распределение жилой квадратуры**

```python
plt.figure(figsize = (13,9))
sns.histplot(df['sqft_living'], bins = 15, color='yellow')
plt.xlabel('Квадратные метры жилой площади')
plt.ylabel('Количество квадратных метров')
plt.title('Распределение жилой площади в метрах');
```

Большинство предложений на рынке недвижимости находится в диапазоне от 500 до 5000 квадратных метров. Наибольшее количество предложений по квадратуре жилой площади находится в диапазоне до в 2000 квадратных метров. Таким образом, наиболее востребованы среди покупателей и арендаторов помещения с небольшой квадратурой.

1.4. **Изучите распределение года постройки**

```python
df['yr_built'].describe()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df['yr_built'], bins = range(1900, 2030, 10), color='purple')
ax.xaxis.set_major_locator(plt.IndexLocator(base = 10, offset = 0))
ax.grid()
for i in ax.patches:
    ax.annotate('%1.0f'%(i.get_height()), (i.get_x() + 2, i.get_height() + 70))
plt.title('Распределение недвижимости по году постройки')
plt.xlabel('Год постройки')
plt.ylabel('Количество объектов недвижимости')
plt.show()
```

Максимальное количество объектов недвижимости находится в диапазоне с 1940 по 2010 годы постройки.

**2.1. Изучите распределение домов от наличия вида на набережную**

```python
data = df['waterfront'].value_counts()

plt.figure(figsize=(17, 7))
plt.pie(data, autopct='%1.1f%%')
plt.legend(['Объекты недвижимости без вида на набережную', 'Объекты недвижимости с видом на набережную']);
```

Большинство объектов недвижимости (99,2%) не имеют вида на набережную.

**2.2. Изучите распределение этажей домов**

```python
counts = df.groupby('floors').size()
colors = 'red', 'purple', 'pink', 'grey', 'orange'
counts.plot(kind='bar', color=colors)
plt.xlabel('Этажность')
plt.ylabel('Количество объектов недвижимости')
plt.title('Распределение объектов недвижимости по этажности')
plt.show()
```

В данной выборке объектов недвижимости преобладают малоэтажные застройки.

**2.3. Изучите распределение состояния домов**

```python
data = df['condition'].value_counts()
fig, ax = plt.subplots(figsize=(15, 10))
data.plot.barh(ax=ax, width=0.8, color='red', align='center', label='Количество')
plt.title('Распределение объектов недвижимости по состоянию')
plt.xlabel('Количество объектов недвижимости')
plt.ylabel('Оценка состояния объектов недвижимости')
plt.show()
```

Большинство объектов недвижимости находятся в хорошем и удовлетворительном состоянии, что является положительным фактором для данного жилого фонда.

**Задача 3 ** Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока. Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
df.info()
```

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix = np.round(corr_matrix, 2)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0 # Заменяем значения корреляции меньше 0.3 на 0

plt.figure(figsize=(17, 13))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
# Диаграмма 1. Влияние года постройки на стоимость объекта недвижимости
plt.figure(figsize=(17, 7))
sns.lineplot(data=df, x='yr_built', y='price', color='purple')
plt.xlabel('год постройки')
plt.ylabel('Стоимость объекта недвижимости (цена в УЕ)')
plt.title('Распределение стоимости объекта недвижимости от года постройки')
plt.show()
```

```python
# Диаграмма 2. Влияние квадратуры жилой площади на стоимость объекта недвижимости

plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='sqft_living', y='price', color='purple')
plt.xlabel('Квадратура жилой площади')
plt.ylabel('Стоимость объекта недвижимости')
plt.title('Зависимость стоимости объекта недвижимости от квадратуры жилой площади')
plt.show()
```

```python
# Диаграмма 3. Влияние количества комнат.
plt.figure(figsize=(17, 7))
sns.lineplot(data=df, x='bedrooms', y='price', color='purple')
plt.xlabel('Количество комнат')
plt.ylabel('Стоимость объектов недвижимости (цена в УЕ)')
plt.title('Распределение стоимости объектов недвижимости в зависимости от количества комнат')
plt.show()
```

```python
# Диаграмма 4. Влияние состояния жилплощади на стоимость объекта недвижимости
plt.figure(figsize=(17, 15))
sns.lineplot(data=df, x='sqft_living', y='price', alpha=0.7, color='purple')
plt.xlabel('Состояние жилой площади')
plt.ylabel('Стоимость недвижимости (цена в УЕ)')
plt.title('Зависимость стоимости объекта недвижимости от состояния жилой площади')
plt.show()
```

```python
# Диаграмма 5. Влияние категории объекта недвижимости на стоимость объекта недвижимости

plt.figure(figsize=(10, 12))

sns.boxplot(x=df['grade'], y=df['price'], whis=0.8, color='purple')

plt.xlabel('Категория объекта недвижимости')
plt.ylabel('Стоимость объекта недвижимости (цена УЕ)')
plt.title('Распределение цен на объекты недвижимости в зависимости от категории недвижимости');
```

ОБЩИЕ ВЫВОДЫ

Ценообразование на объекты недвижимости зависит от многих факторов. Одним из ведущих факторов, повышающих стоимость объекта недвижимости - категория недвижимости (состояние жилого объекта, ремонт и дизайн), соответственно чем выше категория, тем выше стоимость объекта недвижимости. Год постройки также влияет на стоимость: дома до 1940 года могут стоить подороже, также рост цен замечен у домов новой постройки после 2000 года. Распределение стоимости недвижимости в зависимости от квадратуры жилой площади и количества комнат показывает, что увеличение площади жилья и количества комнат (как жилых, так и ванных комнат) приводит к росту стоимости объекта недвижимости.
